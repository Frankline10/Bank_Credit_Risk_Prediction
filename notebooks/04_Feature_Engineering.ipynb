# Phase 4 :- Feature Engineering

```python
# Importing Libraries

import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
# Loading cleaned datasets

account = pd.read_csv('/Users/soujanyaprokashsingha/Documents/Data Analyst_Science/Professional Projects/Datamites Projects/Bank_Credit_Risk_Prediction/data/processed/account_cleaned.csv')

enquiry = pd.read_csv('/Users/soujanyaprokashsingha/Documents/Data Analyst_Science/Professional Projects/Datamites Projects/Bank_Credit_Risk_Prediction/data/processed/enquiry_cleaned.csv')

demo = pd.read_csv('/Users/soujanyaprokashsingha/Documents/Data Analyst_Science/Professional Projects/Datamites Projects/Bank_Credit_Risk_Prediction/data/processed/demo_cleaned.csv')
```

```python
# Converting Date Columns

date_cols_account = [
    'opened_dt',
    'last_paymt_dt',
    'closed_dt',
    'reporting_dt',
    'paymt_str_dt',
    'paymt_end_dt'
]

for col in date_cols_account:
    account[col] = pd.to_datetime(account[col], errors='coerce')
```

```python
# Converting enquiry date to date-time

enquiry['enquiry_dt'] = pd.to_datetime(
    enquiry['enquiry_dt'],
    errors='coerce'
)
```

```python
# using the copy of "Demo" because of keeping the demo unmodified

master_df = demo.copy()
```

# Account based features

```python
# Total accounts per customer

total_accounts = (
    account.groupby('customer_no')
    .size()
    .reset_index(name='total_accounts')
)
```

```python
# Merging Into Master Dataset

master_df = master_df.merge(
    total_accounts,
    on='customer_no',
    how='left'
)
```

```python
# Creating Total Credit Limit

total_credit_limit = (
    account.groupby('customer_no')['creditlimit']
    .sum()
    .reset_index(name='total_credit_limit')
)
```

```python
master_df = master_df.merge(
    total_credit_limit,
    on='customer_no',
    how='left'
)
```

```python
# Create Total Current Balance

total_balance = (
    account.groupby('customer_no')['cur_balance_amt']
    .sum()
    .reset_index(name='total_current_balance')
)
```

```python
master_df = master_df.merge(
    total_balance,
    on='customer_no',
    how='left'
)
```

```python
# Creating Total Past Due Amount

total_past_due = (
    account.groupby('customer_no')['amt_past_due']
    .sum()
    .reset_index(name='total_past_due')
)
```

```python
master_df = master_df.merge(
    total_past_due,
    on='customer_no',
    how='left'
)
```

```python
# Creating and merging Average Credit Limit

avg_credit_limit = (
    account.groupby('customer_no')['creditlimit']
    .mean()
    .reset_index(name='avg_credit_limit')
)
```

```python
master_df = master_df.merge(
    avg_credit_limit,
    on='customer_no',
    how='left'
)
```

```python
# maximum credit limit

max_credit_limit = (
    account.groupby('customer_no')['creditlimit']
    .max()
    .reset_index(name='max_credit_limit')
)
```

```python
master_df = master_df.merge(
    max_credit_limit,
    on='customer_no',
    how='left'
)
```

```python
# Average current balance

avg_balance = (
    account.groupby('customer_no')['cur_balance_amt']
    .mean()
    .reset_index(name='avg_current_balance')
)
```

```python
master_df = master_df.merge(
    avg_balance,
    on='customer_no',
    how='left'
)
```

```python
# Maximum current balance

max_balance = (
    account.groupby('customer_no')['cur_balance_amt']
    .max()
    .reset_index(name='max_current_balance')
)
```

```python
master_df = master_df.merge(
    max_balance,
    on='customer_no',
    how='left'
)
```

```python
# Creating and merging Credit Utilization Ratio

master_df['credit_utilization_ratio'] = (
    master_df['total_current_balance'] /
    master_df['total_credit_limit']
)
```

```python
# Creating and merging average past due

avg_past_due = (
    account.groupby('customer_no')['amt_past_due']
    .mean()
    .reset_index(name='avg_past_due')
)
```

```python
master_df = master_df.merge(
    avg_past_due,
    on='customer_no',
    how='left'
)
```

```python
# Creating and merging maximum past due

max_past_due = (
    account.groupby('customer_no')['amt_past_due']
    .max()
    .reset_index(name='max_past_due')
)
```

```python
master_df = master_df.merge(
    max_past_due,
    on='customer_no',
    how='left'
)
```

```python
# Creating account duration

account['account_age_days'] = (
    account['last_paymt_dt'] -
    account['opened_dt']
).dt.days
```

```python
# Aggregating and merging customer-level age

avg_account_age = (
    account.groupby('customer_no')['account_age_days']
    .mean()
    .reset_index(name='avg_account_age_days')
)
```

```python
master_df = master_df.merge(
    avg_account_age,
    on='customer_no',
    how='left'
)
```

```python
# Creating and merging closed account count

closed_accounts = (
    account.groupby('customer_no')['closed_dt']
    .apply(lambda x: x.notnull().sum())
    .reset_index(name='closed_accounts_count')
)
```

```python
master_df = master_df.merge(
    closed_accounts,
    on='customer_no',
    how='left'
)
```

```python
# Average Interest rate

avg_interest_rate = (
    account.groupby('customer_no')['rateofinterest']
    .mean()
    .reset_index(name='avg_interest_rate')
)
```

```python
master_df = master_df.merge(
    avg_interest_rate,
    on='customer_no',
    how='left'
)
```

```python
# Average payment amount

avg_payment_amount = (
    account.groupby('customer_no')['actualpaymentamount']
    .mean()
    .reset_index(name='avg_payment_amount')
)
```

```python
master_df = master_df.merge(
    avg_payment_amount,
    on='customer_no',
    how='left'
)
```

# Enquiry based features

```python
# Total enquiry count

total_enquiries = (
    enquiry.groupby('customer_no')
    .size()
    .reset_index(name='total_enquiries')
)
```

```python
master_df = master_df.merge(
    total_enquiries,
    on='customer_no',
    how='left'
)
```

```python
# Average enquiry amount

avg_enquiry_amount = (
    enquiry.groupby('customer_no')['enq_amt']
    .mean()
    .reset_index(name='avg_enquiry_amount')
)
```

```python
master_df = master_df.merge(
    avg_enquiry_amount,
    on='customer_no',
    how='left'
)
```

```python
# Maximum enquiry amount

max_enquiry_amount = (
    enquiry.groupby('customer_no')['enq_amt']
    .max()
    .reset_index(name='max_enquiry_amount')
)
```

```python
master_df = master_df.merge(
    max_enquiry_amount,
    on='customer_no',
    how='left'
)
```

```python
# Creating days since enquiry

reference_date = enquiry['enquiry_dt'].max()

enquiry['days_since_enquiry'] = (
    reference_date - enquiry['enquiry_dt']
).dt.days
```

```python
recent_enquiry = (
    enquiry.groupby('customer_no')['days_since_enquiry']
    .min()
    .reset_index(name='recent_enquiry_days')
)
```

```python
master_df = master_df.merge(
    recent_enquiry,
    on='customer_no',
    how='left'
)
```

```python
# Creating enquiry feature since last 90 days

enquiry_90 = enquiry[
    enquiry['days_since_enquiry'] <= 90
]

count_enquiry_90 = (
    enquiry_90.groupby('customer_no')
    .size()
    .reset_index(name='count_enquiry_90')
)
```

```python
master_df = master_df.merge(
    count_enquiry_90,
    on='customer_no',
    how='left'
)
```

```python
# Creating enquiry feature since last 365 days

enquiry_365 = enquiry[
    enquiry['days_since_enquiry'] <= 365
]

count_enquiry_365 = (
    enquiry_365.groupby('customer_no')
    .size()
    .reset_index(name='count_enquiry_365')
)
```

```python
master_df = master_df.merge(
    count_enquiry_365,
    on='customer_no',
    how='left'
)
```

```python
# Most frequent enquiry purpose

most_freq_enquiry = (
    enquiry.groupby('customer_no')['enq_purpose']
    .agg(lambda x: x.mode()[0] if not x.mode().empty else np.nan)
    .reset_index(name='most_freq_enquiry_purpose')
)
```

```python
master_df = master_df.merge(
    most_freq_enquiry,
    on='customer_no',
    how='left'
)
```

# Payment history features

```python
# Payment history length and aggregation

account['payment_history_length'] = (
    account['paymenthistory1']
    .astype(str)
    .apply(len)
)
```

```python
avg_payment_history_length = (
    account.groupby('customer_no')['payment_history_length']
    .mean()
    .reset_index(name='avg_payment_history_length')
)
```

```python
master_df = master_df.merge(
    avg_payment_history_length,
    on='customer_no',
    how='left'
)
```
