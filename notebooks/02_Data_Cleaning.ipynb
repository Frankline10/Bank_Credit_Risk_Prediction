# Phase 2 :- Data Cleaning and Preprocessing

```python
# Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
# Loading all datasets

account_df = pd.read_csv('/Users/soujanyaprokashsingha/Documents/Data Analyst_Science/Professional Projects/Datamites Projects/Bank_Credit_Risk_Prediction/data/raw/Cust_Account.csv')

enquiry_df = pd.read_csv('/Users/soujanyaprokashsingha/Documents/Data Analyst_Science/Professional Projects/Datamites Projects/Bank_Credit_Risk_Prediction/data/raw/Cust_Enquiry.csv')

demo_df = pd.read_csv('/Users/soujanyaprokashsingha/Documents/Data Analyst_Science/Professional Projects/Datamites Projects/Bank_Credit_Risk_Prediction/data/raw/Cust_Demographics.csv')
```

```python
# Creating copies before cleaning

account = account_df.copy()
enquiry = enquiry_df.copy()
demo = demo_df.copy()
```

```python
# Calculating missing percentage for all datasets

missing_percent = (account.isnull().sum()/len(account))*100

missing_percent.sort_values(ascending=False)
```

```python
missing_percent = (enquiry.isnull().sum()/len(enquiry))*100

missing_percent.sort_values(ascending=False)
```

```python
# Removing duplicate columns in each datasets

account.duplicated().sum()

account.drop_duplicates(inplace=True)
```

```python
enquiry.duplicated().sum()

enquiry.drop_duplicates(inplace=True)
```

```python
demo.duplicated().sum()

demo.drop_duplicates(inplace=True)
```

```python
# Standardizing column names in each dataset (Lowercase)

account.columns = account.columns.str.lower()

enquiry.columns = enquiry.columns.str.lower()

demo.columns = demo.columns.str.lower()
```

```python
# Strip all columns

account.columns = account.columns.str.strip()

enquiry.columns = enquiry.columns.str.strip()

demo.columns = demo.columns.str.strip()
```

```python
# Converting date columns to datetime from account dataset

date_cols_account = [
    'opened_dt',
    'last_paymt_dt',
    'closed_dt',
    'reporting_dt',
    'paymt_str_dt',
    'paymt_end_dt'
]

for col in date_cols_account:
    account[col] = pd.to_datetime(account[col], errors='coerce')
```

```python
account.info()
```

```python
# Converting enquiry date

enquiry['enquiry_dt'] = pd.to_datetime(
    enquiry['enquiry_dt'],
    errors='coerce'
)
```

```python
enquiry.info()
```

```python
# converting numeric columns properly

numeric_cols = [
    'high_credit_amt',
    'cur_balance_amt',
    'amt_past_due',
    'creditlimit',
    'cashlimit'
]

for col in numeric_cols:
    account[col] = pd.to_numeric(account[col], errors='coerce')
```

```python
# Negative values

(account[numeric_cols] < 0).sum()
```

```python
# Standardize text cols

account['acct_type'] = account['acct_type'].astype(str).str.lower().str.strip()
```
